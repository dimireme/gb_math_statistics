## Урок 1. Случайные события. Условная вероятность. Формула Байеса. Независимые испытания.

```python
from math import factorial
```

```python
"""число сочетаний из n по k"""
comb = lambda n, k: 0 if k > n else factorial(n) // (factorial(k) * factorial(n-k))

"""число перестановок из n по k"""
perm = lambda n, k: 0 if k > n else factorial(n) // factorial(n-k)
```

**Задача 1. Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.**

Всего исходов вытянуть 4 карты из 52:
$$C_{52}^4 = 270725$$
Исходов вытянуть 4 карты, среди которых не будет туза:
$$C_{48}^4 = 194580$$
Значит искомое число исходов
$$n = C_{52}^4 - C_{48}^4 = 76145$$
**Ответ: 76145**

```python
n = comb(52, 4) - comb(48, 4)
print(n)
```

**Задача 2. Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть: а) два определённых человека? б) три определённых человека?**

а) Всего рассадить 7 человек можно $7!$ способами. Рассадить двоих рядом, без учёта перестановок внутри пары, можно шестью способами. Число перестановок внутри группы из двух человек $A_2^2 = 2$. Рассадить оставшихся 5 человек по 5 местам можно $5!$ способами. В итоге получим 
$$P = \frac{A_2^2\cdot6\cdot5!}{7!}=0.2857$$

б) Рассадить троих рядом, без учёта перестановок внутри группы, можно пятью способами. Число перестановок внутри группы из трёх человек $A_3^3 = 6$. Рассадить оставшихся 4 человека по 4 местам можно $4!$ способами. В итоге получим 
$$P = \frac{A_3^3\cdot5\cdot4!}{7!}=0.1428$$

**Ответ: а) 0.2587 б) 0.1428**

```python
p_a = perm(2,2)*6*factorial(5) / factorial(7)
p_b = perm(3,3)*5*factorial(4) / factorial(7)
print(p_a, p_b)
```

**Задача 3. Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?**

а) Всего возможных исходов вытянуть 3 билета из 60 $C_{60}^3$. Успешных исходов $C_{50}^3$. Искомая вероятность 
$$P = \frac{C_{50}^3}{C_{60}^3} = 0.5728$$

б) Возможных исходов вытянуть 2 билета из 50 изученных $C_{50}^2$. Исходов вытянуть 1 билет из 10 неизученных $C_{10}^1$. Искомая вероятность
$$P = \frac{C_{50}^2\cdot C_{10}^1}{C_{60}^3} = 0.358$$

**Ответ: а) 0.5728 б) 0.358**

```python
p_a = comb(50, 3) / comb(60, 3)
p_b = comb(50, 2) * comb(10, 1) / comb(60, 3)
print(p_a, p_b)
```

**Задача 4. Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?**

По определению, события A и B будут независимыми, если $P(A|B) == P(A)$. Мы знаем что вероятность выпадения чётного числа $P(A) = {1\over2}$. 

$P(A|B) = \frac{P(AB)}{P(B)}$. Вероятность того что выпавшее число больше трёх и чётное $2\over6$, так как под это условие подходят числа 4 и 6 из шести возможных. Вероятность того что выпавшее число будет больше трёх $P(B) = {3\over6}$. Значит $P(A|B) = {2\over3}$.

Так как $P(A|B) \neq P(A)$, то события A и B зависимы.

**Ответ: нет**

**Задача 5. Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ. Какова вероятность, что вы действительно больны ей?**

***Подсказка: используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.***

Пусть A - событие что пациент болен:
$$P(A) = 0.001$$
$$P(\overline A) = 0.999$$

Событие B - тест показал что пациент болен. Это возможно, если пациент болен и прибор показал правильно, либо пациент здоров и тест ошибся: 
$$P(B) = P(A) \cdot P(B|A) + P(\overline A) \cdot P(B|\overline A)$$
$$P(B) = 0.001  \cdot 0.99 + 0.999 \cdot 0.01 = 0.01098$$

В задаче нужно найти $P(A|B)$, то есть что пациент болен при условии что прибор показал что пациент болен.
$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$
$$P(A|B) = \frac{0.99 \cdot 0.001}{0.01098} = 0.090164$$

**Ответ: 0.090164**
